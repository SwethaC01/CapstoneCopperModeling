```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew

class DataPlotter:
    def __init__(self, data):
        self.data = data

    def skews_plot(self, values1, values2):
        
        for v1, v2 in zip(values1, values2):
            plt.figure(figsize=(18, 6))
            
            # Boxplot before transformation
            plt.subplot(1, 3, 1)
            sns.boxplot(x=v1, data=self.data, color="rebeccapurple")
            plt.title(f'Boxplot - Before Transformation')

            # Histogram before transformation
            plt.subplot(1, 3, 2)
            sns.histplot(self.data[v1], kde=True, bins=30, color="aqua")
            plt.title(f'Histogram - Before Transformation\nSkewness: {skew(self.data[v1])}')

            # Violinplot before transformation
            plt.subplot(1, 3, 3)
            sns.violinplot(x=v1, data=self.data, color="saddlebrown")
            plt.title(f'Violinplot - Before Transformation')

            plt.tight_layout()
            plt.show()
            
            plt.figure(figsize=(18, 6))

            # Boxplot after transformation
            plt.subplot(1, 3, 1)
            sns.boxplot(x=v2, data=self.data, color="mediumvioletred")
            plt.title(f'Boxplot - After Transformation')

            # Histogram after transformation
            plt.subplot(1, 3, 2)
            sns.histplot(self.data[v2], kde=True, bins=30, color="royalblue")
            plt.title(f'Histogram - After Transformation\nSkewness: {skew(self.data[v2])}')

            # Violinplot after transformation
            plt.subplot(1, 3, 3)
            sns.violinplot(x=v2, data=self.data, color="navy")
            plt.title(f'Violinplot - After Transformation')

            plt.tight_layout()
            plt.show()

plotter = DataPlotter(datapre)
plotter.skews_plot(values1, values2)
```

## **Outlier Handling**

```python
#Finding the IQR
percentile25 = datapre['selling_price'].quantile(0.25)
percentile75 = datapre['selling_price'].quantile(0.75)
iqr=percentile75 - percentile25
print(iqr)
print(percentile25)
print(percentile75)

#Finding upper and lower limit
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
print(f"upper bound of {upper_limit}")
print(f"lower bound of {lower_limit}")
```

```python
# For all columns
def iqr_lower_bound(datapre, col):
    Q1 = datapre[col].quantile(0.25)
    Q3= datapre[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - (1.5*IQR)
    return lower_bound
    
def iqr_upper_bound(datapre, col):
    Q1 = datapre[col].quantile(0.25)
    Q3= datapre[col].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + (1.5*IQR)
    return upper_bound
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor, ExtraTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from xgboost import XGBRegressor

class RegressionModel:
    def __init__(self, df):
        self.df = df
        self.preprocess_data()
        self.split_data()
        self.initialize_models()
        
    def preprocess_data(self):
        self.x_reg = self.df[['quantity_tons_log','status_en', 'item_type_en', 'application', 'thickness_log', 'width', 'country', 'customer', 'product_ref']]
        self.y_reg = self.df['selling_price_log']
    
    def split_data(self):
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.x_reg, self.y_reg, test_size=0.3, random_state=5)
        print(f'Train shape: {self.X_train.shape}')
        print(f'Test shape: {self.X_test.shape}')
        print(f'Train labels shape: {self.y_train.shape}')
        print(f'Test labels shape: {self.y_test.shape}')
    
    def initialize_models(self):
        self.models = {
            'LinearRegression': LinearRegression(),
            'DecisionTreeRegressor': DecisionTreeRegressor(),
            'RandomForestRegressor': RandomForestRegressor(),
            'GradientBoostingRegressor': GradientBoostingRegressor(),
            'ExtraTreeRegressor': ExtraTreeRegressor(),
            'XGBRegressor': XGBRegressor()
        }
    
    def evaluate_model(self, model_name, model):
        model.fit(self.X_train, self.y_train)
        y_pred = model.predict(self.X_test)
        mse = mean_squared_error(self.y_test, y_pred)
        r2 = r2_score(self.y_test, y_pred)
        mae = mean_absolute_error(self.y_test, y_pred)
        
        return {
            'Model_Name': model_name,
            'Model': model,
            'Mean_Squared_Error': mse,
            'R_Squared': r2,
            'Mean_Absolute_Error': mae
        }
    
    def all_models(self):
        results = [self.evaluate_model(model_name, model) for model_name, model in self.models.items()]
        self.model_score_df = pd.DataFrame(results)
        return self.model_score_df.drop(columns=['Model'])
    
    def get_best_model(self, metric='R_Squared'):
        if self.model_score_df is None:
            raise ValueError("Model scores are not available. Please run all_models() first.")

        best_model_row = self.model_score_df.loc[self.model_score_df[metric].idxmax()]
        return best_model_row['Model']

model = RegressionModel(df1)
results = model.all_models()

print("\n**Evaluation Metrics for All Algorithms:**\n")
print(results.to_string(index=False))

best_model = model.get_best_model()
print("\n**Best Model Based on R_Squared:**\n")
print(results.loc[results['R_Squared'].idxmax()].to_string(index=False))
```
